# Project 1: Explanatory Data Analysis & Data Presentation (Movies Dataset)

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies...

2. We are going to __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

Here we are going to copy the most specific columns from our dataframe that we are going to work with

```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
             "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
#Calculate the Profit
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)

#Calculate the Highest Return on Investment
df_best["return"] = df["revenue_musd"].div(df["budget_musd"])

df_best.T
```

Change the column names for good data presentation

```python
df_best.columns = ["Image", "Title", "budget", "Revenue", "Votes",
                  "Average Rating", "Poplarity", "Profit", "ROI"]
```

```python
df_best.set_index("Title", inplace = True)
```

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
subset = df_best.iloc[:5,:2]
subset
```

```python
df_best["budget"].fillna(0, inplace = True)
df_best["Votes"].fillna(0, inplace = True)
```

```python
df_best.sort_values(by = "Average Rating", ascending = False).head()
```

As we can see the most highly rated film, but it seems to be rather naive if we have the top rated films with an average rating of 10. However, there is a problem when it comes to films where there is only one vote. And of course that's not really significant, so we should add an extra feature and an extra filter. We should therefore be able to filter films based on the number of votes and only get films for which we are missing at least 10 votes.

```python
df_best.sort_values(by = "ROI", ascending = False).head()
```

In fact, we have the same problem with ROI.
So if we think that all films come back from above to below, then we get films with a very high return. And the reason is because the budget is so close to zero for this film. So it's not really meaningful. And we should be able to filter out films that are at least on budget.

```python
df_best.loc[df_best.budget >= 5].sort_values(by = "ROI", ascending = False).head()
```

## Are Franchises more successful?

4.We are goign to __Analyze__ the Dataset and __find out whether Franchises (Movies that belong to a collection) are more successful than stand-alone movies__ in terms of:

- mean revenue
- median Return on Investment
- mean budget raised
- mean popularity
- mean rating

```python
df["belongs_to_collection"]
```

```python
df["Franchise"] = df.belongs_to_collection.notna()
```

```python
df["Franchise"]
```

```python
df.Franchise.value_counts()
```

__Franchise vs. Stand-alone: Average Revenue__

```python
df.groupby("Franchise").revenue_musd.mean()
```

__Franchise vs. Stand-alone: Return on Investment / Profitability (median)__

```python
df["ROI"] = df.revenue_musd.div(df.budget_musd)
```

```python
df.groupby("Franchise").ROI.median()
```

__Franchise vs. Stand-alone: Average Budget__

```python
df.groupby("Franchise").budget_musd.mean()
```

__Franchise vs. Stand-alone: Average Popularity__

```python
df.groupby("Franchise").popularity.mean()
```

__Franchise vs. Stand-alone: Average Rating__

```python
df.groupby("Franchise").vote_average.mean()
```

```python
#to summarize...
df.groupby("Franchise").agg({"budget_musd":"mean",
                            "revenue_musd": "mean",
                            "vote_average": "mean",
                            "popularity": "mean", "ROI":"median",
                            "vote_count": "mean"}).T
```
