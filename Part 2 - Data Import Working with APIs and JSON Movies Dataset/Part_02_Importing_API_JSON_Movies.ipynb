##  Importing and Saving the Movies Dataset 

5. __API-Request__ (movie module): in this time, we are going to Load all available information for the movies with movie id = [__299534, 19995, 140607, 299536, 597, 135397, 420818, 24428, 168259, 99861, 284054, 12445, 181808, 330457, 351286, 109445, 321612, 260513__] into a Pandas DataFrame and __save the dataset in a local json file__.

```python
import json
```

```python
movie_id = [299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 
            260513]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}' 
```

```python
json_list = []
for movie in movie_id :
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df1 = pd.DataFrame(json_list)
```

```python
df1.loc[:, ["title", "id", "revenue", "genres", "belongs_to_collection", "runtime"]].sort_values(by ="revenue", ascending = True).head()
```
